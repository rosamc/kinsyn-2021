```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib as mpl
import json
```

```python
green="#c6cdad"
blue="#d3dbe4"
red="#e3cbc8"
mpl.rcParams["font.size"]=14
def tint_ax(ax):
    xmin,xmax=ax.get_xlim()
    ymin,ymax=ax.get_ylim()
    if xmax>0:
        ax.fill_between([0,xmax],ymax,color=green,zorder=1)
    if xmin<0:
        ax.fill_between([xmin,0],ymax,color=blue,zorder=1)
        ax.fill_between([xmin,0],ymin,color=red,zorder=1)
    ax.set_xlim(xmin,xmax)
    ax.set_ylim(ymin,ymax)
    return ax
```

```python
#Fig 2 main
filename='./files/ashapes_0_4_1000_10_run0903.json'
ashapes=json.load(open(filename,'r'))
data1=ashapes["difAD_samebnp_step11"]["fc1=10_fc2=10"]
data2=ashapes["difAD_samebnp_step12"]["fc1=10_fc2=10"]
data3=ashapes["difAD_samebnp"]["fc1=10_fc2=10"]
fig,axes=plt.subplots(3,1,figsize=(4*0.75,10*0.8))
titles=["$k_1-k_1$","$k_1-k_2$","any-any"]
xlim0=-3
xlim1=12
ylim1=12
for d, data in enumerate([data1,data2,data3]):
    ax=axes[d]
    ax.plot(data["x"],data["y"],color='k')
    ax.axvline(x=0,color="gray",linestyle="--")
    ax.axhline(y=0,color="gray",linestyle="--")
    ax.set_xlim(xlim0,xlim1)
    ax.set_ylim(-1,ylim1)
    ax=tint_ax(ax)
    ax.set_xlabel("$S_{A,B}$",fontsize=16)
    ax.set_ylabel("$S_{B,A}$",fontsize=16,labelpad=-5)
    ax.set_xticks([0,5,10])
    ax.set_title(titles[d])
plt.tight_layout()
#plt.savefig("figs/boundariesFig2.pdf")
plt.show()
```
